```python
#locate the topic and the link
url = 'https://www.who.int/health-topics/'


response = requests.get(url)
response

body = response.content
soup = BeautifulSoup(body, 'html.parser')

topics = []
for tag in soup.find_all('a', class_="link-container table"):
    topic_name = tag.find('p', class_="heading text-underline").get_text()
    topic_link = tag['href']
    topics.append({
        'name': topic_name,
        'links':topic_link
    })

#collect data for each topic or link
data = []
for topic in topics:
    print(f"Scraping content for:{topic['name']}")

#fetch topic page through link
    links = topic['links'] #e.g 'https://www.who.int/health-topics/alcohol#tab=tab_1'
    topic_response = requests.get(links)
    soup = BeautifulSoup(topic_response.text, 'html.parser')

    #locate the tags for the.....
    contents = soup.find_all('div', class_="sf_colsOut tabContent")

    #create a heading
    headings = {
        'Overview': 'N/A',
        'Impacts': 'N/A',
        'WHO_Response': 'N/A'
        }
    
    if len(contents) > 0:
        headings['Overview']= contents[0].get_text(strip = True)
    if len (contents) > 1:
        headings['Impacts'] = contents[1].get_text(strip = True)
    if len (contents) > 2:
        headings['WHO_Response'] = contents[2].get_text(strip = True)
    
    
    #Append data
    data.append({
        "Health Topic" : topic['name'],
        "Links" : topic['links'],
        "Overviews" : headings['Overview'],
        "Impacts" : headings ['Impacts'],
        "Response" : headings ['WHO_Response']   
    })
```

```python
df = pd.DataFrame(data)
```
